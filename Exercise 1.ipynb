```python
unit_list = []
power_list = []
# matrix_of_conversion = [[1,1000, 1000000], [0.001, 1, 1000], [0.000001, 0.001, 1]]
# print(matrix_of_conversion[2][2])
converted_power_list = []
converted_units_list = []

number_of_pair_values = input("Please indicate the number of power values you want to enter: ")
target_unit = input("Please indicate the target value you want to convert: ")

for i in range(1,int(number_of_pair_values)+1):
    
    power = int(input("Please type the power values:"))
    power_list.append(power)
    
    unit = str(input("Please type the unit: "))
    
    if unit.casefold() == "kw":
        unit_list.append(unit)
        
    elif unit.casefold() == "gw":
        unit_list.append(unit)
        
    elif unit.casefold() == "mw":
        unit_list.append(unit)
        
    else:
        print("Error. Please note that is not a valid unit")

    if unit.casefold() != target_unit.casefold():
        if unit.casefold() == "mw" and target_unit.casefold() == "kw":
            converted_power_list.append(power*1000)
            
        if unit.casefold() == "gw" and target_unit.casefold() == "kw":
            converted_power_list.append(power*1000000)
            
        if unit.casefold() == "kw" and target_unit.casefold() == "mw":
            converted_power_list.append(power/1000)
            
        if unit.casefold() == "gw" and target_unit.casefold() == "mw":
            converted_power_list.append(power*1000)
        
        if unit.casefold() == "kw" and target_unit.casefold() == "gw":
            converted_power_list.append(power/1000000)
            
        if unit.casefold() == "mw" and target_unit.casefold() == "gw":
            converted_power_list.append(power/1000)
            
    else:
        converted_power_list.append(power)
        
    converted_units_list.append(target_unit.casefold())
                
pair_of_converted_values = [item for sublist in zip(converted_power_list, converted_units_list) for item in sublist]


print(power_list, unit_list, converted_power_list, pair_of_converted_values)
```
